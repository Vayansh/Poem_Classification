# Preprocessing Step

```python
import re
```

```python
import nltk
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords
```

```python
stop_words=set(stopwords.words())
```

```python
def simple_model_preprocessing(text):
  text = re.sub("@[A-Za-z0-9_]+"," ",text)
  text = re.sub("#[A-Za-z0-9_]+"," ",text)
  text = re.sub(r'[^\w\r]',' ',text)

  text = ' '.join([item for item in text.split() if item not in stop_words])
  text = ' '.join([item for item in text.split() if item.isalpha()])
  return text
```

```python
df['Poem'] = df['Poem'].apply(simple_model_preprocessing)
```

```python
test_df['Poem'] = test_df['Poem'].apply(simple_model_preprocessing)
```

# Applying Simple Machine Learning model and Count Vectorizer

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import LogisticRegression
```

```python
X_train, y_train = df['Poem'], df['Genre']
```

```python
X_test, y_test = test_df['Poem'], test_df['Genre']
```

```python
pipe = Pipeline([('vect', CountVectorizer(stop_words='english', max_df=100)),

                ('model', LogisticRegression(solver = 'liblinear',max_iter = 5000))])

# Fitting the model

model1 = pipe.fit(df['Poem'], df['Genre'])

prediction = model1.predict(test_df['Poem'])
print('model score = ',pipe.score(df['Poem'], df['Genre']))
print("accuracy: {}%".format(round(accuracy_score(test_df['Genre'], prediction)*100,2)))
```

```python
pipe = Pipeline([('vect', CountVectorizer()),

                 ('tfidf', TfidfTransformer()),

                ('model', LogisticRegression(max_iter = 5000))])

# Fitting the model

model1 = pipe.fit(df['Poem'], df['Genre'])

prediction = model1.predict(test_df['Poem'])

print("accuracy: {}%".format(round(accuracy_score(test_df['Genre'], prediction)*100,2)))
```

```python
from sklearn.tree import DecisionTreeClassifier

pipe = Pipeline([('vect', CountVectorizer()),
                ('tfidf', TfidfTransformer()),
                ('model', DecisionTreeClassifier(max_depth = 75))])
model1 = pipe.fit(df['Poem'], df['Genre'])

prediction = model1.predict(test_df['Poem'])

print("accuracy: {}%".format(round(accuracy_score(test_df['Genre'], prediction)*100,2)))
```
